### Monthly Income

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.boxplot(y=df['monthly_income'])
plt.title("Monthly Income Before Outlier Treatment")
plt.show()
```

```python
# Nmuerical ke through outlier check kar rhe h
Q1 = df['monthly_income'].quantile(0.25)
Q3 = df['monthly_income'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("Lower Bound:",lower_bound)
print("Upper Bound:",upper_bound)
```

```python
# mein ismai CAP ka use kiya h 
import numpy as np

df['monthly_income'] = np.where(
    df['monthly_income'] < lower_bound,
    lower_bound,
    df['monthly_income']
)

df['monthly_income'] = np.where(
    df['monthly_income'] > upper_bound,
    upper_bound,
    df['monthly_income']
)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['monthly_income'])
plt.title("Monthly Income After Outlier Treatment")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
import warnings

warnings.filterwarnings('ignore')

# 1. Load and clean the base data
df = pd.read_csv('personal_loan_dataset.csv').dropna(subset=['loan_status'])
df = df.drop(columns=['name'])

# Define feature columns
cat_cols = ['city_tier', 'employment_type']
num_cols = [c for c in df.columns if c not in cat_cols + ['loan_status', 'approved_amount', 'interest_rate']]

# 2. Impute Features (X)
# Filling missing feature values with median/most frequent
num_imputer = SimpleImputer(strategy='median')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Handle categorical missing and encoding
for col in cat_cols:
    df[col] = df[col].fillna('Unknown')

label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# 3. Model 1: Classification (Loan Status)
X = df[num_cols + cat_cols]
y_status = df['loan_status']
clf_model = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42).fit(X, y_status)

# 4. Model 2: Regression (Approved Amount)
# CRITICAL FIX: Only take rows where status=1 AND amount is NOT NaN
amt_df = df[(df['loan_status'] == 1) & (df['approved_amount'].notna())]
X_amt = amt_df[num_cols + cat_cols]
y_amt = amt_df['approved_amount']
amt_model = RandomForestRegressor(n_estimators=200, random_state=42).fit(X_amt, y_amt)

# 5. Model 3: Regression (Interest Rate)
# CRITICAL FIX: Only take rows where status=1 AND interest_rate is NOT NaN
int_df = df[(df['loan_status'] == 1) & (df['interest_rate'].notna())]
X_int = int_df[num_cols + cat_cols]
y_int = int_df['interest_rate']
int_model = RandomForestRegressor(n_estimators=200, random_state=42).fit(X_int, y_int)

# 6. Accurate Prediction Function
def get_accurate_prediction(input_data):
    input_df = pd.DataFrame([input_data])
    
    # Preprocess
    for col, le in label_encoders.items():
        val = input_df[col].iloc[0]
        # Handle unseen categories in real-time input
        if val not in le.classes_:
            input_df[col] = le.transform([le.classes_[0]]) # Default to first class
        else:
            input_df[col] = le.transform([val])
            
    input_df = input_df[num_cols + cat_cols]
    
    # Predictions
    prob = clf_model.predict_proba(input_df)[0][1]
    is_approved = clf_model.predict(input_df)[0]
    
    # Logic for Risk
    risk = "Low Risk" if (input_data['cibil_score'] > 750 and prob > 0.8) else \
           "Medium Risk" if (input_data['cibil_score'] > 650) else "High Risk"

    if is_approved == 1:
        pred_amt = amt_model.predict(input_df)[0]
        pred_int = int_model.predict(input_df)[0]
        print(f"--- Approval is accepted ---")
        print(f"Customer Risk : {risk}")
        print(f"Loan Decision : Approved")
        print(f"Approval Probability : {prob:.2f}")
        print(f"Approval Amount : ₹ {int(pred_amt):,}")
        print(f"Interest Rate : {pred_int:.1f}%\n")
    else:
        print(f"--- Approval is denied ---")
        print(f"Customer Risk : {risk}")
        print(f"Loan Decision : Rejected")
        print(f"Approval Probability : {prob:.2f}")
        print(f"Approval Amount : NA")
        print(f"Interest Rate : NA\n")

# --- TEST ---
test_user = {
    'age': 34, 'city_tier': 'Metro', 'employment_type': 'Salaried',
    'monthly_income': 80477, 'other_income': 18814, 'bank_avg_balance_6m': 18950,
    'job_tenure_months': 11, 'cibil_score': 795, 'past_defaults': 3,
    'inquiries_6m': 3, 'existing_emi': 28938, 'dti': 9.085,
    'loan_amount_requested': 128002, 'tenure_requested_months': 16,
    'insurer_present': 1, 'co_applicant': 0
}
get_accurate_prediction(test_user)
```
